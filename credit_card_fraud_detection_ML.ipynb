# Data Set Information

The dataset used in this project contains transactions made by credit cards in September 2013 by European cardholders. It includes transactions that occurred over two days, with the following characteristics:

- Total transactions:284,807
- Fraudulent transactions: 492 (approximately 0.172% of all transactions)

### Key Points
1. Highly Unbalanced Data:
   - The dataset is highly imbalanced, with fraudulent transactions accounting for only 0.172% of all transactions.
   
2. Features:
   - The dataset contains only numerical input variables.
   - Most features (`V1`, `V2`, ..., `V28`) are the principal components obtained through PCA transformation.
   - Two features are not PCA-transformed:
     - `Time`: The seconds elapsed between each transaction and the first transaction in the dataset.
     - `Amount`: The transaction amount, useful for example-dependent cost-sensitive learning.
   - `Class`: The response variable, where:
     - `1` indicates a fraudulent transaction.
     - `0` indicates a legitimate transaction.

3. Confidentiality Constraints:
   - Due to confidentiality, the original feature names and additional background information about the data are unavailable.

4. Class Imbalance Challenges:
   - Traditional metrics like confusion matrix accuracy are not meaningful due to the class imbalance.
   - Recommended Metric:Area Under the Precision-Recall Curve (AUPRC).

This dataset poses a significant challenge in identifying fraudulent transactions, requiring careful handling of the class imbalance and appropriate evaluation metrics to ensure model reliability.

```python
#Split
X = df.drop(columns=['Class'], axis=1)
y = df['Class']
```

```python
# standard scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_scaler = sc.fit_transform(X)
```

```python
# Model Training
# train test split
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score
x_train, x_test, y_train, y_test = train_test_split(x_scaler, y, test_size=0.25, random_state=42, stratify=y)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
# training
model.fit(x_train, y_train)
# testing
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
```

```python
#random forest classifier
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
# training
model.fit(x_train, y_train)
# testing
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
```

```python
# hint - use combination of over sampling and under sampling
# balance the class with equal distribution
from imblearn.over_sampling import SMOTE
over_sample = SMOTE()
x_smote, y_smote = over_sample.fit_resample(x_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_jobs=-1)
# training
model.fit(x_smote, y_smote)
# testing
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
```
